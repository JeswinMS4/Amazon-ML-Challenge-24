```python
from PIL import Image
import requests
from io import BytesIO
import pandas as pd
import re

# Define the function to load images from URLs
def load_image(image_url):
    response = requests.get(image_url)
    return Image.open(BytesIO(response.content)).convert('RGB')

# Define the function to extract weight values with units
def extract_weights(text):
    if not isinstance(text, str):  # Ensure text is a string
        return []

    # Regex patterns for various weight units
    patterns = {
        'gram': r'(\d+(\.\d+)?\s*gram|\d+(\.\d+)?\s*g|gram\s*\d+(\.\d+)?)',
        'kilogram': r'(\d+(\.\d+)?\s*kilogram|\d+(\.\d+)?\s*kg|kilogram\s*\d+(\.\d+)?)',
        'microgram': r'(\d+(\.\d+)?\s*microgram|\d+(\.\d+)?\s*µg|microgram\s*\d+(\.\d+)?)',
        'milligram': r'(\d+(\.\d+)?\s*milligram|\d+(\.\d+)?\s*mg|milligram\s*\d+(\.\d+)?)',
        'ounce': r'(\d+(\.\d+)?\s*ounce|\d+(\.\d+)?\s*oz|ounce\s*\d+(\.\d+)?)',
        'pound': r'(\d+(\.\d+)?\s*pound|\d+(\.\d+)?\s*lb|pound\s*\d+(\.\d+)?)',
        'ton': r'(\d+(\.\d+)?\s*ton|\d+(\.\d+)?\s*t|ton\s*\d+(\.\d+)?)'
    }

    weights = []
    for unit, pattern in patterns.items():
        matches = re.findall(pattern, text, re.IGNORECASE)
        for match in matches:
            weight = match[0].strip()
            weights.append((weight, unit))

    return weights

# Define a function to get the prediction from the model
def get_prediction(image_url, question):
    # Load the image
    image = load_image(image_url)

    # Prepare the messages for the model
    msgs = [{'role': 'user', 'content': [image, question]}]

    # Query the model
    answer = model.chat(
        image=None,
        msgs=msgs,
        tokenizer=tokenizer
    )

    # Extract weight values and units from the model's answer
    weights = extract_weights(answer.strip())

    # Format the result as a string (if multiple results, you may want to handle this differently)
    if weights:
        ans = ', '.join([f"{weight} {unit}" for weight, unit in weights])
        print(ans)
        return ans
    else:
        ans = ""
        print(ans)
        return ans

# Load the DataFrame
df = pd.read_csv("/content/nan_predictions.csv")

# Filter the DataFrame to include only rows where entity_name is 'item_weight'
df_weight = df[df['entity_name'] == 'item_weight']
df_weight = df_weight.sample(n=500, random_state=1)  # random_state is set for reproducibility

# Define the question
question = 'What is the weight of the item? Provide only one value in standard format and add the corresponding units at the end: gram, kilogram, microgram, milligram, ounce, pound, ton.'

# Create a new column for model predictions
df_weight['model_prediction'] = None

# Iterate over the filtered DataFrame and update the new column with predictions
for index, row in df_weight.iterrows():
    if pd.notna(row['image_link']):  # Ensure there is an image link
        prediction = get_prediction(row['image_link'], question)
        df_weight.at[index, 'model_prediction'] = prediction

# Save the updated DataFrame if needed
df_weight.to_csv("updated_weight.csv", index=False)

# Display a sample of the updated DataFrame
print(df_weight.sample(5))
```

```python
import pandas as pd
import re

# Define a mapping of weight units to their full names
unit_fullname_weight = {
    'g': 'gram',
    'kg': 'kilogram',
    'mg': 'milligram',
    'µg': 'microgram',
    'ounce': 'ounce',
    'lb': 'pound',
    'pound': 'pound',
    'ton': 'ton'
}

# Function to extract weight values with units
def extract_weights(text):
    if not isinstance(text, str):  # Ensure text is a string
        return []

    # Regex patterns for various weight units
    patterns = {
        'gram': r'(\d+(\.\d+)?\s*g|g\s*\d+(\.\d+)?)',
        'kilogram': r'(\d+(\.\d+)?\s*kg|kg\s*\d+(\.\d+)?)',
        'microgram': r'(\d+(\.\d+)?\s*µg|µg\s*\d+(\.\d+)?)',
        'milligram': r'(\d+(\.\d+)?\s*mg|mg\s*\d+(\.\d+)?)',
        'ounce': r'(\d+(\.\d+)?\s*ounce|ounce\s*\d+(\.\d+)?)',
        'pound': r'(\d+(\.\d+)?\s*lb|lb\s*\d+(\.\d+)?)',
        'ton': r'(\d+(\.\d+)?\s*ton|ton\s*\d+(\.\d+)?)'
    }

    weights = []
    for unit, pattern in patterns.items():
        matches = re.findall(pattern, text, re.IGNORECASE)
        for match in matches:
            weight = match[0].strip()
            weights.append((weight, unit))

    return weights

# Function to extract the numeric value from the weight string
def extract_numeric_value(weight_str):
    numeric_part = re.findall(r"[-+]?\d*\.\d+|\d+", weight_str)
    return float(numeric_part[0]) if numeric_part else 0

# Function to format the largest weight value
def format_largest_weight_value(weight_list):
    if not weight_list:
        return ""

    # Find the tuple with the largest numeric value
    largest_value_tuple = max(weight_list, key=lambda x: extract_numeric_value(x[0]))

    # Get the numeric value and the unit
    largest_value = extract_numeric_value(largest_value_tuple[0])
    largest_unit = largest_value_tuple[1]

    # Format the result with the full unit name
    formatted_value = f"{int(largest_value)}" if largest_value.is_integer() else f"{largest_value}"
    return f"{formatted_value} {unit_fullname_weight.get(largest_unit.lower(), largest_unit)}"

# Function to get the largest formatted weight value
def get_largest_formatted_weight(prediction):
    weights = extract_weights(prediction)
    return format_largest_weight_value(weights)

# Load the DataFrame
df_weights = pd.read_csv("/kaggle/input/nan-predictions/nan_predictions.csv")

# Filter to include only rows where entity_name is 'item_weight'
df_weights = df_weights[df_weights['entity_name'] == 'item_weight']

# Create a new column for the largest formatted weight prediction
df_weights['prediction'] = df_weights['model_prediction'].apply(get_largest_formatted_weight)

# Display the updated DataFrame
print(df_weights[['model_prediction', 'prediction']])
```

```python
import pandas as pd

def replace_predictions(test_meta, width_df):
    """
    Replaces the 'prediction' values in the 'test_meta' DataFrame with those from 'width_df'
    based on common 'image_link' values.

    Parameters:
    - test_meta (pd.DataFrame): DataFrame containing the 'image_link' and 'prediction' columns.
    - width_df (pd.DataFrame): DataFrame containing the 'image_link' and 'prediction' columns.

    Returns:
    - pd.DataFrame: Updated 'test_meta' DataFrame with replaced 'prediction' values.
    """

    # Ensure the columns exist in both DataFrames
    if 'image_link' not in test_meta.columns or 'prediction' not in test_meta.columns:
        raise ValueError("test_meta DataFrame must contain 'image_link' and 'prediction' columns")

    if 'image_link' not in width_df.columns or 'prediction' not in width_df.columns:
        raise ValueError("width_df DataFrame must contain 'image_link' and 'prediction' columns")

    # Merge the DataFrames on 'image_link', using 'width_df' values for 'prediction'
    updated_df = test_meta.merge(width_df[['image_link', 'prediction']], on='image_link', how='left', suffixes=('', '_new'))

    # Replace the old 'prediction' values with the new ones where available
    updated_df['prediction'] = updated_df['prediction_new'].combine_first(updated_df['prediction'])

    # Drop the temporary 'prediction_new' column
    updated_df.drop(columns=['prediction_new'], inplace=True)

    return updated_df

# Example usage
# test_meta = pd.DataFrame({
#     'image_link': ['link1', 'link2', 'link3'],
#     'prediction': [0.1, 0.2, 0.3]
# })

# width_df = pd.DataFrame({
#     'image_link': ['link2', 'link3'],
#     'prediction': [0.5, 0.6]
# })

updated_test_meta = replace_predictions(test_meta, df_weight)
print(updated_test_meta)
```
